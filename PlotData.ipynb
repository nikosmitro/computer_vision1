```
import re

numbers = []
with open("granada.txt", "r") as f:
    for line in f:
        numbers += re.findall("t\d{4}", line)
        print(numbers)
print(numbers)
```

```
temp_gr = []
with open("granada.txt", "r") as f:
    for line in f:
        if 't' in line:
            start = line.index('t')
            temp_gr.append(float(line[start+1:start+5]))
#temp_gr = float(temp_gr)  
print(temp_gr)
print(len(temp_gr))
```

```
hum_gr = []
with open("granada.txt", "r") as f:
    for line in f:
        if 'h' in line:
            start = line.index('h')
            hum_gr.append(float(line[start+1:start+6]))

#hum_gr = float(hum_gr)              
print(hum_gr)
print(len(hum_gr))
```

```
temp_rho = []
hum_rho = []
with open("rhodes.txt", "r") as f:
    for line in f:
        if 't' in line:
            start = line.index('t')
            temp_rho.append(float(line[start+1:start+6]))
        if 'h' in line:
            start = line.index('h')
            hum_rho.append(float(line[start+1:start+6]))

#temp_rho = float(temp_rho)  
print(temp_rho)
print(len(temp_rho))
#hum_rho = float(hum_rho)  
print(hum_rho)
print(len(hum_rho))
```

```
temp_ven = []
hum_ven = []
with open("venice.txt", "r") as f:
    for line in f:
        if 't' in line:
            start = line.index('t')
            temp_ven.append(float(line[start+1:start+5]))
        if 'h' in line:
            start = line.index('h')
            hum_ven.append(float(line[start+1:start+6]))
#temp_ven = float(temp_ven)           
print(temp_ven)
print(len(temp_ven))
#hum_ven = float(hum_ven)  
print(hum_ven)
print(len(hum_ven))
```

```
temp_ton = []
hum_ton = []
with open("tonsberg.txt", "r") as f:
    for line in f:
        if 't' in line:
            start = line.index('t')
            temp_ton.append(float(line[start+1:start+6]))
        if 'h' in line:
            start = line.index('h')
            hum_ton.append(float(line[start+1:start+6]))

#temp_ton = float(temp_ton)
print(temp_ton)
print(len(temp_ton))
#hum_ton = float(hum_ton)
print(hum_ton)
print(len(hum_ton))
```

```
rssi = pd.read_excel("/content/timestamps.xlsx")

tons = list((rssi['Tonsberg']))
ven = list((rssi['Venice']))
rho = list(rssi['Rhodes'])
gra = list(rssi['Granada'])
print(tons)
```

```
tons1=[]
for t in tons:
    tons1.append(t.strftime("%H:%M"))
print(tons1)
```

```


plt.figure(figsize=(20,10))
plt.style.use('seaborn')
plt.plot(tons1[1:],temp_gr,'darkorange',label='Granada')
plt.plot(tons1[1:],temp_rho,'maroon',label='Rhodes')
plt.plot(tons1[1:],temp_ton,'coral',label='Tonsberg')
plt.plot(tons1[1:],temp_ven,'red',label='Venice')
plt.xticks(tons1[1:], tons1[1:], rotation =45)
plt.title(r'Temperature on 5-6 January 2022', fontsize=16)
plt.xlabel(r'Time (hh:mm) (GMT+00:00)', fontsize=16)
plt.ylabel(r'Temperature (°C)', fontsize=16)
#plt.xlabel("Time (secs)")
#plt.ylabel("RSSI (dB)")
plt.legend(loc=4, prop={'size': 16})
plt.show()
```

```
plt.figure(figsize=(20,10))
plt.style.use('seaborn')
plt.plot(tons1[1:],hum_gr,'teal',label='Granada')
plt.plot(tons1[1:],hum_rho,'skyblue',label='Rhodes')
plt.plot(tons1[1:],hum_ton,'navy',label='Tonsberg')
plt.plot(tons1[1:],hum_ven,'blue',label='Venice')
plt.xticks(tons1[1:], tons1[1:], rotation =45)
plt.title(r'Relative Humidity on 5-6 January 2022', fontsize=16)
plt.xlabel(r'Time (hh:mm) (GMT+00:00)', fontsize=16)
plt.ylabel(r'Relative Humidity (\%)', fontsize=16)
#plt.xlabel("Time (secs)")
#plt.ylabel("RSSI (dB)")
plt.legend(loc=4, prop={'size': 16})
plt.show()
```
